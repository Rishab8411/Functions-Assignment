1. What is the difference between a function and a method in Python?
    - A function is a block of code that performs a specific task and can be called independently. A method is similar but is associated with an object and is called using that object.

```python
# Example:
def greet():  # Function
    print("Hello!")

class Person:
    def say_hello(self):  # Method
        print("Hello from a method!")

obj = Person()
obj.say_hello()  # Calling method
greet()  # Calling function
```

2. Explain the concept of function arguments and parameters in Python?
    - Parameters are the variables in a function definition.

    - Arguments are the values passed to the function during execution.

```python
# Example:
def add(a, b):  # a and b are parameters
    return a + b

print(add(5, 10))  # 5 and 10 are arguments, Output: 15
```

3. What are the different ways to define and call a function in Python?
   -  Python supports multiple ways to define and call functions:

```python
# Normal function:
def greet():
    print("Hello!")
greet()

# Function with parameters:
def add(a, b):
    return a + b
print(add(3, 5))

# Lambda function:
square = lambda x: x * x
print(square(4))

# Recursive function:
def factorial(n):
    return 1 if n == 0 else n * factorial(n - 1)
print(factorial(5))
```

5. What are iterators in Python and how do they differ from iterables?
    - Iterable: Any object that can return an iterator (e.g., lists, tuples).
    - Iterator: An object that can be iterated using __iter__() and __next__().

```python
# Example:
my_list = [1, 2, 3]
iterator = iter(my_list)  # Creating an iterator
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
```

6. Explain the concept of generators in Python and how they are defined?
    - Generators use yield to return values one by one instead of storing them in memory.

```python
# Example:
def count_up(n):
    for i in range(n):
        yield i

gen = count_up(5)
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
```

7. What are the advantages of using generators over regular functions?
    - Saves memory as values are generated one by one.

    - Lazy evaluation, meaning values are produced only when needed.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
    - map() applies a function to all elements in an iterable.
    - reduce() reduces an iterable to a single value using a function.
    - filter() filters elements based on a condition.

```python
# Example:
from functools import reduce

numbers = [1, 2, 3, 4]

mapped = list(map(lambda x: x*2, numbers))  # [2, 4, 6, 8]
filtered = list(filter(lambda x: x%2 == 0, numbers))  # [2, 4]
reduced = reduce(lambda x, y: x + y, numbers)  # 10

print(mapped, filtered, reduced)
```
